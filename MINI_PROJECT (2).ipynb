```python
# target=[]
# images=[]
# flat_data=[]

# DATADIR='/content/images'
# CATEGORIES=['Ball','Tree','Icecream','Toy','Shoes']
# for category in CATEGORIES:
#   class_num=CATEGORIES.index(category) #Label Encoding
#   path=os.path.join(DATADIR,category)
#   for img in os.listdir(path):
#     img_array=imread(os.path.join(path,img))
#     # print(img_array.shape)
#     # plt.imshow(img_array)
#     # break
#     img_resized=resize(img_array,(180,180,3))
#     flat_data.append(img_resized.flatten())
#     images.append(img_resized)
#     target.append(class_num)

# flat_data=np.array(flat_data)
# target=np.array(target)
# images=np.array(images)

import os
import numpy as np
from PIL import Image
from skimage.transform import resize

DATADIR = '/content/images'
CATEGORIES = ['Ball', 'Tree', 'Icecream', 'Toy', 'Shoes']

target = []
images = []
flat_data = []

for category in CATEGORIES:
    class_num = CATEGORIES.index(category)  # Label Encoding
    path = os.path.join(DATADIR, category)
    for filename in os.listdir(path):
        img_path = os.path.join(path, filename)
        if os.path.isfile(img_path):
            img = Image.open(img_path)
            img_array = np.array(img)
            img_resized = resize(img_array, (180, 180, 3))
            flat_data.append(img_resized.flatten())
            images.append(img_resized)
            target.append(class_num)

flat_data = np.array(flat_data)
target = np.array(target)
images = np.array(images)
```

```python
unique,count=np.unique(target,return_counts=True)
plt.bar(CATEGORIES,count)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(flat_data,target,test_size=0.3,random_state=210)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn import svm
param_grid=[
    {'C':[1,10,100,1000],'kernel':['Linear']},
    {'C':[1,10,100,1000],'gamma':[0.001,0.0001],'kernel':['rbf']},
]

svc=svm.SVC(probability=True)
clf= GridSearchCV(svc,param_grid)
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(y_pred,y_test)
```

```python
confusion_matrix(y_pred,y_test)
```
